```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns 
```

# Lectura de datos

```python
titanic = pd.read_csv('titanic.csv')
titanic.head(10)
```

# Descripción de variables

```python
# Indica cuántos registros se tienen en total, cuáles tienen valores nulos y cuáles no tienen valores nulos, 
# así como el tipo de dato de cada columna.

titanic.info()

# Identifica el tipo de variables presentes en el dataset, 
# y en una celda de texto indica cuáles variables son categóricas y cuáles numéricas

print(titanic.dtypes)

print('Variables categóricas: ')
print('Pclass - Survived - Sex, Embarked')
print('Variables numéricas: ')
print('Age - SibSp - Parch - Fare')

```

# Exploracion y analisis de dato

```python
# Muestra un resumen estadístico de las variables cuantitativas.
titanic.describe()
```

```python
# Analiza la información y contesta las siguientes preguntas ¿Cuál es la tarifa más barata? ¿Cuál es el promedio de edad?
print('La tarifa más barata es de', titanic['Fare'].min())
print('El promedio de edad es de', titanic['Age'].mean())
```

```python
# Muestra un resumen estadístico de las variables cualitativas, incluyendo survived y pclass. 
# Puedes cambiar su tipo de dato usando la función de pandas astype
titanic['Survived'] = titanic['Survived'].astype('category')
titanic['Pclass'] = titanic['Pclass'].astype('category')
titanic.describe(include='category')
```

```python
# Indica la distribución de frecuencias de la columna embarked, así como la gráfica de barras correspondiente.
titanic['Embarked'].value_counts()
titanic['Embarked'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Muestra los datos del pasajero con id = 149.
titanic.loc[149]
```

```python
# Mostrar los pasajeros que iban en primera clase, ordenadas alfabéticamente por nombre.
titanic[titanic['Pclass'] == 1].sort_values('Name')
```

```python
# Agrega el código para indicar cuánto es el total que se pagó sumando la tarifa de todos los pasajeros.
print(titanic['Fare'].sum())
```

```python
# Crea un subconjunto de los datos de los que sobrevivieron y otro de los que no sobrevivieron. Usa agrupación.
sobrevivientes = titanic.groupby('Survived')
sobrevivientes.get_group(1)
```

```python
sobrevivientes.get_group(0)
```
