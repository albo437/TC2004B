```python
import pandas as pd
from sklearn import datasets
from sklearn import metrics
```

```python
dataset = datasets.load_breast_cancer()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target
```

```python
y = df.target
x = df.drop('target', axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

algoritmo = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p=2)
algoritmo.fit(x_train, y_train)

y_pred = algoritmo.predict(x_test)

print('Precisión del modelo K-NN:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

algoritmo = GaussianNB()
algoritmo.fit(x_train, y_train)
y_pred = algoritmo.predict(x_test)

print('Precisión del modelo Naive Bayes:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.svm import SVC

algoritmo = SVC(kernel='linear')
algoritmo.fit(x_train, y_train)
y_pred = algoritmo.predict(x_test)

print('Precisión del modelo SVM:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

algoritmo = LogisticRegression(max_iter=10000)
algoritmo.fit(x_train, y_train)
y_pred = algoritmo.predict(x_test)

print('Precisión del modelo Regresión Logística:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

algoritmo = DecisionTreeClassifier(criterion='entropy')
algoritmo.fit(x_train, y_train)
y_pred = algoritmo.predict(x_test)

print('Precisión del modelo Árbol de Decisión:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier

algoritmo = MLPClassifier(hidden_layer_sizes=(8,8,8), activation='relu', solver='adam', max_iter=500)
algoritmo.fit(x_train, y_train)
y_pred = algoritmo.predict(x_test)

print('Precisión del modelo Redes Neuronales:')
print(metrics.accuracy_score(y_test, y_pred))
```

En general todos lo modelos obtienen un desempeño similar, el unico que suele generar bajos mas bajos es K-NN, pero cuando le va bien le va igual de bien que a los demas.
